```python
 treinamentos = [{'treinamento':'Scrum','moedas':30},
{'treinamento':'Data Science','moedas':40},
{'treinamento':'Gesto de Projetos','moedas':50},
{'treinamento':'Marketing','moedas':30},
{'treinamento':'Cloud','moedas':20},
{'treinamento':'Blockchain','moedas':10},
{'treinamento':'Python','moedas':30}]

pedidos = ['Scrum','Data Science','Gesto de Projetos','Marketing','Cloud','Python','Python','Python',
'Scrum','Data Science','Gesto de Projetos','Marketing','Data Science','Gesto de Projetos',
'Python','Marketing','Data Science','Gesto de Projetos','Data Science','Gesto de Projetos','Data Science']
```

```python
pedidos_unicos = {}
lista_pedidos_unicos = []
qtd = 0

# Acessando os dicionarios na lista treinamentos
for t in treinamentos:
  for k, v in t.items():
    # acessando os valores de cada pedido
    for p in pedidos:
      # alinhando o pedido com o cadastro na lista treinamentos para puxar valor 
      if k == 'treinamento' and v == p:
        qtd = pedidos.count(p)
        # verificando duplicatas na hora de adc valores  lista final
        if p in pedidos_unicos.values():
          break
        else:
          pedidos_unicos['pedido'] = p
          pedidos_unicos['qtd'] = qtd
          pedidos_unicos['valor'] = t['moedas']
          lista_pedidos_unicos.append(pedidos_unicos.copy())

# realizando o sorted da lista de dicionrios nova, em ordem decrescente da maior quantidade de pedidos para a menor
lista_pedidos_unicos_sorted = sorted(lista_pedidos_unicos, key=lambda lista_pedidos_unicos: lista_pedidos_unicos['qtd'], reverse=True)

soma = 0
cursos_viaveis = []

for p in lista_pedidos_unicos_sorted:
  for u, w in p.items():
    if u == 'valor':
      if soma >= 150:
        soma_final = soma - w
      # s faz a soma do valor do curso e coloca ele na lista de cursos viveis uma vez que a soma no tenha passado 150
      else:
        soma += w
        # aqui  necessrio entrar de novo na lista p porm acessando os values da key pedidos para puxar os nomes
        for k, v in p.items():
          if k == 'pedido':
            cursos_viaveis.append(v) 
print(f'Os cursos viveis realizao so: {cursos_viaveis}, no valor de {soma}')
```
