### Lista dentro de for

```python
l1 = [0,2,6,4,8,1,3]
l2 = [3,2,1,7,8,2,5]
l3 = []

for i, valor in enumerate(l1):
  l3.append(l1[i] + l2[i])
print(l3)
```

```python
# ele para o for no final da lista menor
l1 = [0,2,6,4,8,1,3]
l2 = [3,2,1,7]
l3 = []

for v1, v2 in zip(l1, l2):
  l3.append(v1 + v2)
print(l3)
```
