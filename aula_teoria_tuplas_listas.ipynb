# Tuplas
<br> TUPLAS SÃO SEPARADAS POR VIRGULAS DENTRO DE PARENTESES
<br> Variáveis compostas imutáveis
<br> guarda mais de uma informação de uma vez

```python
var = ('lanche','batata','refri')
```

```python
var[0]
```

```python
print(var)
```

```python
print(var[2])
```

```python
print(var[0:2])
```

```python
print(var[1:])
```

```python
print(type(var[:2]))
```

```python
print(type(var[-1]))
```

```python
tupla = (1,'aaaa',True)
```

```python
t1 = (0,6,4,2,8)
t2 = (1,3,3,5,3,5,7)
```

```python
len(t1)
```

```python
sorted(t1)
```

```python
type(sorted(t1))
```

```python
# conta quantas vezes apareceu o 3 dentro de t2
t2.count(3)
```

```python
# verifica a posição do 3 dentro de t2 (se tiver mais ocorrencias, mostra a primeira)
t2.index(3)
```

```python
# vê a posição a partir de uma posição
t2.index(3,3)
```

```python
# isso aqui CONCATENA as tuplas, e não soma
t2 + t1
```

```python
t3 = (2,6,'batata',5,'o prego ta enferrujado')

for i in t3:
  print(i)
print('☻')
```

```python
t4 = (2, 6, 4, 5, 7)
soma = 0

for i in t4:
  soma += i
print(soma)
```

```python
t5 = ('lanche','batata','refri')

for ind, valor in enumerate(t5):
  print('item {}: é um: {}'.format(ind,valor))
print('☻')
```
