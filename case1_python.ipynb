```python
area = [200,'null',720,1500,1000,275,'null',1200,2400,'null']
altura = ['null',30,40,30,10,25,33,'null',12,20]
largura = [20,20,'null',50,100,'null',30,100,200,10]
```

```python
for i,valor in enumerate(area):
  if area[i] == 'null':
    area[i] = altura[i] * largura[i]
  else:
    area[i] = area[i]

for i,valor in enumerate(altura):
  if altura[i] == 'null':
    altura[i] = area[i] / largura[i]
  else:
    area[i] = area[i]

for i,valor in enumerate(largura):
  if largura[i] == 'null':
    largura[i] = area[i] / altura[i]
  else:
    area[i] = area[i]

print(area)
print(altura)
print(largura)
```

```python
area = [200, 600, 720, 1500, 1000, 275, 990, 1200, 2400, 200]
altura = [10.0, 30, 40, 30, 10, 25, 33, 12.0, 12, 20]
largura = [20, 20, 18.0, 50, 100, 11.0, 30, 100, 200, 20]
```

```python
mediaAltura = sum(altura)/len(altura)
mediaLargura = sum(largura)/len(largura)
print(mediaAltura)
print(mediaLargura)
```

```python
comparar = []

for i,valor in enumerate(altura):
  if altura[i] > largura[i]:
    resultado = 1
  else:
    resultado = 0
  comparar.append(resultado)

print(comparar)
print(100*sum(comparar)/len(comparar),'% das vezes a altura  maior que a largura')
```
